```python
import matplotlib.pyplot as plt

# test 3
x = [50, 66, 75, 80, 90, 95, 98, 99, 100]
express = [316, 323, 327, 329, 337, 345, 414, 440, 632]
express_pm2 = [270 ,280 ,287 ,292 ,304 ,315 ,326 ,337 ,400]
gin = [62 ,64 ,66 ,67 ,72 ,76 ,86 ,96 ,198]

#plot y vs x
plt.xticks(x, x)
plt.plot(x, express, label="express")
plt.plot(x, express_pm2, label="express pm2")
plt.plot(x, gin, label="gin")
plt.legend(loc='upper left')

#set title and x, y - axes labels
plt.title('Test 3')
plt.xlabel('% of requests served within a certain time (ms)')
plt.ylabel('Time (in ms)')

plt.rcParams["figure.figsize"] = (15, 5)
#show plot to user
plt.show()
```
